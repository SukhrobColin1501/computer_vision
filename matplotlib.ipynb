```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array( [1, 2, 3, 4])
y = np.array([10,25, 37, 41])
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.array( [1, 2, 3, 4])
y = np.array([4, 7, 37, 12])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Styled Line Example")
plt.show()
```

Multiple lines in One Graph with Numpy and Matplotlib

```python
x = np.array( [1, 2, 3, 4])
y1 = np.array([7,9, 13, 17])
y2 = np.array([3,5, 7, 11])

plt.plot(x, y1, label='Line 1', color='red', marker='s')
plt.plot(x, y2, label='Line 2', color='blue', linestyle='--', marker='o')
plt.title('Multible Lines Example')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.show()
```

Bar Chart with Numpy and Maplotlib

```python
x = np.array(['Salary', 'Expense', 'Charity', 'Food'])
y = np.array([5, 9, 1, 7])

plt.bar(x, y, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Scatter Plot with Numpy and Matplotib


```python
x = np.array([2, 4, 6, 8, 10])
y = np.array([5, 7, 9, 11, 8])

plt.scatter(x, y, color='blue', s=100)
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Histogram with Numpy and Matplotlip

```python
data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10 ])

plt.hist(data, bins=5, color="skyblue", edgecolor='black')
plt.title("Histogram Example")
plt.xlabel('Values')
plt.ylabel('Count')
plt.show()
```

Subplots(Multiple Graphs Together)

```python
x = np.array( [1, 2, 3, 4])
y1 = np.array([7,9, 13, 17])
y2 = np.array([3,5, 7, 11])

plt.figure(figsize=(8, 4)) # width(wide) | 4 => height(tall)

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title('Plot 2')

plt.tight_layout()
plt.show()
```
